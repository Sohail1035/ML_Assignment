```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.cluster as cluster  # Make sure this line is included
import time

# Define the plot_clusters function
def plot_clusters(data, algorithm, args, kwds):
    start_time = time.time()
    labels = algorithm(*args, **kwds).fit_predict(data)
    end_time = time.time()
    
    palette = sns.color_palette('Set3', np.unique(labels).max() + 1)
    colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in labels]
    
    plt.scatter(data.T[0], data.T[1], c=colors, alpha=0.4, s=3, linewidths=1)
    plt.title(f'Clusters found by {str(algorithm.__name__)}\n'
              f'Selected parameters {str(kwds)} {str(args)}', fontsize=11)
    plt.text(-2.0, -2.6, f'Clustering took {end_time - start_time:.3f} s', fontsize=11)
    plt.show()

# Load the data
data = np.load('data.npy')

# Now call the function with KMeans
plot_clusters(data, cluster.KMeans, (), {'n_clusters': 3, 'random_state': 42})
```

# TASK 4: Writing a short technical report analysing the “Clustering Results”

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Load data
data = np.load('data.npy')

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(data)
kmeans_centroids = kmeans.cluster_centers_

# Compute intra-cluster distances
kmeans_intra_distances = []
for i in range(kmeans.n_clusters):
    cluster_points = data[kmeans_labels == i]
    distance = np.mean([np.linalg.norm(point - kmeans_centroids[i]) for point in cluster_points])
    kmeans_intra_distances.append(distance)

# Compute inter-cluster distances
kmeans_inter_distances = []
for i in range(kmeans.n_clusters):
    for j in range(i + 1, kmeans.n_clusters):
        dist = np.linalg.norm(kmeans_centroids[i] - kmeans_centroids[j])
        kmeans_inter_distances.append(dist)

# Plot
plt.figure(figsize=(12, 6))

# Intra-cluster distances
plt.subplot(1, 2, 1)
plt.bar(range(len(kmeans_intra_distances)), kmeans_intra_distances, color='skyblue')
plt.title('KMeans Intra-cluster Distances')
plt.xlabel('Cluster')
plt.ylabel('Distance')

# Inter-cluster distances
plt.subplot(1, 2, 2)
plt.bar(range(len(kmeans_inter_distances)), kmeans_inter_distances, color='salmon')
plt.title('KMeans Inter-cluster Distances')
plt.xlabel('Pair of Clusters')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import numpy as np

# Load data
data = np.load('data.npy')

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data)

# Compute intra-cluster distances
dbscan_unique_labels = np.unique(dbscan_labels)
dbscan_intra_distances = []
for label in dbscan_unique_labels:
    if label == -1:
        continue  # Skip noise
    cluster_points = data[dbscan_labels == label]
    if len(cluster_points) > 1:
        distance = np.mean([np.linalg.norm(point - np.mean(cluster_points, axis=0)) for point in cluster_points])
        dbscan_intra_distances.append(distance)

# Compute inter-cluster distances
dbscan_inter_distances = []
for i in range(len(dbscan_unique_labels)):
    for j in range(i + 1, len(dbscan_unique_labels)):
        if dbscan_unique_labels[i] == -1 or dbscan_unique_labels[j] == -1:
            continue  # Skip noise
        cluster_i = data[dbscan_labels == dbscan_unique_labels[i]]
        cluster_j = data[dbscan_labels == dbscan_unique_labels[j]]
        dist = np.mean([np.linalg.norm(point_i - point_j) for point_i in cluster_i for point_j in cluster_j])
        dbscan_inter_distances.append(dist)

# Plot
plt.figure(figsize=(12, 6))

# Intra-cluster distances
plt.subplot(1, 2, 1)
plt.bar(range(len(dbscan_intra_distances)), dbscan_intra_distances, color='skyblue')
plt.title('DBSCAN Intra-cluster Distances')
plt.xlabel('Cluster')
plt.ylabel('Distance')

# Inter-cluster distances
plt.subplot(1, 2, 2)
plt.bar(range(len(dbscan_inter_distances)), dbscan_inter_distances, color='salmon')
plt.title('DBSCAN Inter-cluster Distances')
plt.xlabel('Pair of Clusters')
plt.ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import pdist, squareform

def calculate_distances(data, labels):
    unique_labels = np.unique(labels)
    
    intra_cluster_distances = []
    inter_cluster_distances = []
    cluster_centers = []
    
    for label in unique_labels:
        if label == -1:
            # Skip noise for DBSCAN/HDBSCAN (-1 is used to label noise points)
            continue
        
        # Extract points belonging to the current cluster
        cluster_points = data[labels == label]
        # Calculate intra-cluster distances (distance between points in the same cluster)
        intra_distances = pdist(cluster_points)
        intra_cluster_distances.append(np.mean(intra_distances))
        
        # Calculate the cluster center
        cluster_center = np.mean(cluster_points, axis=0)
        cluster_centers.append(cluster_center)
    
    # Calculate inter-cluster distances (distance between different cluster centers)
    if len(cluster_centers) > 1:
        inter_distances = pdist(np.array(cluster_centers))
        inter_cluster_distances = np.mean(inter_distances)
    else:
        inter_cluster_distances = 0  # In case there's only one cluster
    
    return np.mean(intra_cluster_distances), inter_cluster_distances
```

```python
import seaborn as sns
import sklearn.cluster as cluster
import hdbscan

def analyze_cluster_distances(data, labels, algorithm_name):
    intra_dist, inter_dist = calculate_distances(data, labels)
    
    print(f"{algorithm_name} - Intra-cluster distance: {intra_dist:.2f}, Inter-cluster distance: {inter_dist:.2f}")
    
    return intra_dist, inter_dist

def plot_distances(distances):
    algorithms = ['KMeans', 'DBSCAN', 'HDBSCAN', 'Birch']
    intra_dists = [d[0] for d in distances]
    inter_dists = [d[1] for d in distances]
    
    plt.figure(figsize=(10, 6))
    
    bar_width = 0.35
    index = np.arange(len(algorithms))
    
    plt.bar(index, intra_dists, bar_width, label='Intra-cluster Distance')
    plt.bar(index + bar_width, inter_dists, bar_width, label='Inter-cluster Distance')
    
    plt.xlabel('Clustering Algorithm')
    plt.ylabel('Distance')
    plt.title('Intra-cluster vs. Inter-cluster Distance')
    plt.xticks(index + bar_width / 2, algorithms)
    plt.legend()
    
    plt.show()

# Load the data
data = np.load('data.npy')

# Run clustering algorithms and calculate distances
distances = []

# KMeans
kmeans_labels = cluster.KMeans(n_clusters=3, random_state=42).fit_predict(data)
distances.append(analyze_cluster_distances(data, kmeans_labels, 'KMeans'))

# DBSCAN
dbscan_labels = cluster.DBSCAN(eps=0.5, min_samples=5).fit_predict(data)
distances.append(analyze_cluster_distances(data, dbscan_labels, 'DBSCAN'))

# HDBSCAN
hdbscan_labels = hdbscan.HDBSCAN(min_cluster_size=10, min_samples=5).fit_predict(data)
distances.append(analyze_cluster_distances(data, hdbscan_labels, 'HDBSCAN'))

# Birch
birch_labels = cluster.Birch(n_clusters=3, threshold=0.5).fit_predict(data)
distances.append(analyze_cluster_distances(data, birch_labels, 'Birch'))

# Plot the distances
plot_distances(distances)
```
